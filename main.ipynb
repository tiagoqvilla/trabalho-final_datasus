```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('sinan_tuberculose.csv', sep=';')
df.head()
```

```python
# Remove colunas irrelevantes
df.drop(columns=['identificador_pessoa2', 'data_extracao','teste_molecular_rapido', 'teste_sensibilidade', 'baciloscopia_escarro','cultura'], inplace=True)

df['data_diagnostico'] = pd.to_datetime(df['data_diagnostico'], format='%m/%Y')
df['data_nascimento'] = pd.to_datetime(df['data_nascimento'], format='%m/%Y')
df['idade'] = df['data_diagnostico'].dt.year - df['data_nascimento'].dt.year - ((df['data_diagnostico'].dt.month < df['data_nascimento'].dt.month) | ((df['data_diagnostico'].dt.month == df['data_nascimento'].dt.month) & (df['data_diagnostico'].dt.day < df['data_nascimento'].dt.day)))

df
```

```python
df.replace({'SIM': 1, 'NÃO': 0, 'IGNORADO': 0}, inplace=True)
df.fillna(0, inplace=True)
df
```

```python
from scipy.stats import chi2_contingency#teste quiquadrado para correlacao
#h0,Sob H0:Oij=Eij para todas as células, no caso correlacao = 0
#h1, correlacao > 0
def test_1(df, column):
    contingency_table = pd.crosstab(df['forma'], df[column])
    res= chi2_contingency(contingency_table)
    return res


results = {}
columns = ['agravos_aids', 'agravos_alcool', 'agravos_diabetes', 'agravos_tabagismo']

for column in columns:
    res = test_1(df, column)
    results[column] = {
        'X^2': res.statistic,
        'p_value': res.pvalue,
    }

results
#em todos os casos p-value < 0.05, rejeita-se h0
```

```python
#testando todos os agarvos juntos
contingency_table = pd.crosstab(df['forma'], [df['agravos_aids'], df['agravos_alcool'], df['agravos_diabetes'], df['agravos_tabagismo']])

res = chi2_contingency(contingency_table)

print(res.statistic)
print(res.pvalue)
```

```python
total = len(df)
pulmonar = len(df[df['forma'] == 'PULMONAR'])
extrapulmonar = len(df[df['forma'] == 'EXTRAPULMONAR'])
both_cases = len(df[df['forma'] == 'PULMONAR + EXTRAPULMONAR'])

# Pirori
P_Pulmonar = (pulmonar + both_cases) / total
P_Extrapulmonar = (extrapulmonar + both_cases) / total

# Verossimilhanca
P_Extrapulmonar_given_Pulmonar = both_cases / (pulmonar+ both_cases)
P_Pulmonar_given_Extrapulmonar = both_cases / (extrapulmonar + both_cases)

# Teorema de bayes
Posteriori_1 = (P_Extrapulmonar_given_Pulmonar * P_Pulmonar) / P_Extrapulmonar
Posteriori_2 = (P_Pulmonar_given_Extrapulmonar * P_Extrapulmonar) / P_Pulmonar

print("Probabilidade de ter pulmonar dado que tem extrapulmonar: ",Posteriori_1)
print("Probabilidade de ter extrapulmonar dado que tem pulmonar: ",Posteriori_2)
```

```python
pulmonar_tb = df[df['forma'].str.contains('PULMONAR')]

agravos = ['agravos_aids', 'agravos_alcool', 'agravos_diabetes', 'agravos_tabagismo']
probabilities = {}

for agravo in agravos:
    extrapulmonar_with_agravo = pulmonar_tb[pulmonar_tb[agravo] == 1]['forma'].str.contains('EXTRAPULMONAR').sum()
    total_with_agravo = pulmonar_tb[pulmonar_tb[agravo] == 1].shape[0]
    probability = extrapulmonar_with_agravo / total_with_agravo if total_with_agravo != 0 else 0
    probabilities[agravo] = probability



probabilities
```
